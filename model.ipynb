```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("supermarket_sales .csv")
```

```python
df.info()
```

```python
df.drop(['Invoice ID'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.columns
```

```python
df['Branch'].unique()
```

```python
df['Branch']=df['Branch'].map({'A':0,'B':1,'C':2})
```

```python
df['City'].unique()
df['City']=df['City'].map({'Yangon':0, 'Naypyitaw':1, 'Mandalay':2})
```

```python
df['Customer type'].unique()
df['Customer type']=df['Customer type'].map({'Member':0, 'Normal':1})
```

```python
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
```

```python
df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
encoding = LabelEncoder()
df['Product line'] = encoding.fit_transform(df['Product line'])
df['Payment'] = encoding.fit_transform(df['Payment'])
```

```python
df.head()
```

```python
import pandas as pd

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year from 'Date' column
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Convert 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Extract hour from 'Time' column
df['Hour'] = df['Time'].dt.hour

# Now df contains 'Month', 'Year', and 'Hour' columns with extracted information
```

```python
df.drop(['Date','Time'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
#Check missing values
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
X=df.drop(['Rating'],axis=1)
y=df['Rating'] 
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression;
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

```python

import pickle
with open('linear_regression_model.pkl', 'wb') as f:
    pickle.dump(model, f)

```
